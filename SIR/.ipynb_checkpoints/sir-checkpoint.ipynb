# Modeling the Spread of a Epidemic

#### Initialize Modules

```python
import numpy as np
import scipy
from scipy.integrate import solve_bvp
from matplotlib import pyplot as plt
```

#### Initialize Variables

```python
def initialize_all(a,b,y0,h):
    """Given an initial and final time a and b, with y(a)=y0, and step size h,
    return several things.
    X: an array from a to b with n elements, where n is the number of steps from a to b.
    Y: an empty array of size (n, y.size), Y[0]=y0.
    h: the step size.
    n: the number of steps to be taken.
    """
    n = int((b-a)/h+1)
    X = np.linspace(a, b, n)
    if isinstance(y0, np.ndarray):
        Y = np.empty((n, y0.size))
    else:
        Y = np.empty(n)
    Y[0] = y0
    return X, Y, h, int(n)
```

### RK4:
#### Rungeâ€“Kutta

```python
def RK4(f,X,Y,h,n):
    for i in range(n-1):
        rk1 = np.array(f(X[i], Y[i]))
        rk2 = f(X[i] + h*.5, Y[i] + (h*.5) *rk1)
        rk3 = f(X[i] + h*.5, Y[i] + (h*.5) *rk2)
        rk4 = f(X[i+1], Y[i] + h*rk3)
        Y[i+1] = Y[i] + (h/6) * (rk1 + 2*rk2 + 2*rk3 + rk4)
    return(Y)
```

## Generic Infection Problem

```python
S_0 = 1 - 6.25 * 1e-7
I_0 = 6.25 * 1e-7
R_0 = 0
starting = np.array([S_0,I_0,R_0])

def ode_f(X,Y):
    return np.array([-1/2*Y[1]*Y[0], .5*Y[1]*Y[0]-.25*Y[1], .25*Y[1]])

h_ = .1
a_ = 0
b_ = 100

plt.title("Infection Phase Model")

ZZ1 = initialize_all(a_, b_, starting, h_)
returned1 = RK4(ode_f, ZZ1[0], ZZ1[1], ZZ1[2], ZZ1[3])
plt.plot(ZZ1[0], returned1[:,0], label="Susceptible")
plt.plot(ZZ1[0], returned1[:,1], label="Infected")
plt.plot(ZZ1[0], returned1[:,2], label="Recovered")    
plt.legend(loc="best")
plt.show()
```

## Modeling a Three Million Population with Five Infected

```python
S_0 = (3e6-5)/3e6
I_0 = 5/3e6
R_0 = 0/3e6
starting = np.array([S_0,I_0,R_0])

def ode_f(X,Y):
    return np.array([-Y[1]*Y[0], Y[1]*Y[0]-(1/3)*Y[1], (1/3)*Y[1]])

h_ = .1
a_ = 0
b_ = 50

plt.title("Infection Phase Model")

ZZ1 = initialize_all(a_, b_, starting, h_)
returned1 = RK4(ode_f, ZZ1[0], ZZ1[1], ZZ1[2], ZZ1[3])
plt.plot(ZZ1[0], returned1[:,0], label="Susceptible")
plt.plot(ZZ1[0], returned1[:,1], label="Infected")
plt.plot(ZZ1[0], returned1[:,2], label="Recovered")    
plt.legend(loc="best")
plt.show()

productive = int(3e6-3e6*max(returned1[:,1]))

print("There are " + str(productive) + " productive people.")
```

## Modeling a Three Million Population with Five Infected:
#### Different Possible Contacts (SIR)

```python
S_0 = (3e6-5)/3e6
I_0 = 5/3e6
R_0 = 0/3e6
starting = np.array([S_0,I_0,R_0])

def ode_f(X,Y):
    return np.array([-(3/10)*Y[1]*Y[0], (3/10)*Y[1]*Y[0]-(1/4)*Y[1], (1/4)*Y[1]])

h_ = .1
a_ = 0
b_ = 500

plt.title("Infection Phase Model")

ZZ1 = initialize_all(a_, b_, starting, h_)
returned1 = RK4(ode_f, ZZ1[0], ZZ1[1], ZZ1[2], ZZ1[3])
plt.plot(ZZ1[0], returned1[:,0], label="Susceptible")
plt.plot(ZZ1[0], returned1[:,1], label="Infected")
plt.plot(ZZ1[0], returned1[:,2], label="Recovered")    
plt.legend(loc="best")
plt.xlabel("Days (T)")
plt.ylabel("Percentage of Population")
plt.show()

productive = int(3e6-3e6*max(returned1[:,1]))

print("There are " + str(productive) + " productive people.")
```

## Modeling a Population with Deaths and Births:
#### The case of influenza and mutating diseases (SIS)

```python
S_0 = (7e9-1000)/7e9
I_0 = 1000/7e9
R_0 = 0/7e9

mu_ = .0001
gamma_ = .1
beta_ = .5
ef_ = 1/50
starting = np.array([S_0,I_0,R_0])

def ode_f(X,Y):
    return np.array([ ef_*Y[2] + mu_*(1-Y[0]) -beta_*Y[1]*Y[0], 
                     (beta_)*Y[1]*Y[0]-(gamma_+mu_)*Y[1],
                     -(ef_+mu_)*Y[2] + gamma_*Y[1]
                    ])


h_ = .1
a_ = 0
b_ = 500

plt.title("Infection Phase Model")

ZZ1 = initialize_all(a_, b_, starting, h_)
returned1 = RK4(ode_f, ZZ1[0], ZZ1[1], ZZ1[2], ZZ1[3])
plt.plot(ZZ1[0], returned1[:,0], label="Susceptible")
plt.plot(ZZ1[0], returned1[:,1], label="Infected")
plt.plot(ZZ1[0], returned1[:,2], label="Recovered")    
plt.legend(loc="best")
plt.xlabel("Days (T)")
plt.ylabel("Percentage of Population")
plt.show()

productive = int(7e9-7e9*max(returned1[:,1]))

print("There are " + str(productive) + " productive people at the peak.")
print("or " + str(productive/7e9) + " of humanity")
```

## Modeling a Population with Deaths and Births:
#### The case of measles and mutating diseases (SIS)

```python
epsilon, lbc, rbc = .1, 1, - 1/3

def ode(x , y):
    return np.array([y[1] , (1/epsilon) * (y[0] - y[0] * y[1])])

def bcs(ya, yb):
    BCa = np.array([ya[0] - lbc]) # 1 Boundary condition on the left
    BCb = np.array([yb[0] - rbc]) # 1 Boundary condition on the right
    # The return values will be 0s when the boundary conditions are met exactly
    return np.hstack([BCa, BCb])

X = np.linspace(-1, 1, 200)
y = np.array([-1/3, -4/3]).reshape((-1,1))*np.ones((2, len(X)))
solution = solve_bvp(ode, bcs, X, y)
y_plot = solution.sol(X)[0]
plt.plot(X, y_plot)
plt.show()
```

```python
be_0 = 1575
be_1 = 1
en_ = 0.01
lam = .0279
mu_ = .02

h_ = .1
a_ = 0
b_ = 1

beta = lambda t: be_0*(1+be_1*np.cos(2*np.pi*t))

def ode(t, y):
    length = len(y[0])
    return np.array([mu_-beta(t)*y[0]*y[2],beta(t)*y[0]*y[2]-y[1]/lam,
                      y[1]/lam-y[2]/en_,0*np.zeros(length),0*np.zeros(length),0*np.zeros(length)])    
# Note the extra 0's on the end, this keeps the C from changing.
 
def bcs(ya, yb):
    BCa = ya[0:3] - ya[3:] # 1 Boundary condition on the left
    BCb = yb[0:3] - yb[3:] # 1 Boundary condition on the right
    return np.hstack([BCa, BCb])
T = np.linspace(0, 1, 100)
Y = np.array([.07, .0001, .0001, .08, .0001, .0001]).reshape((-1,1))*np.ones((6,len(T)))

solution = solve_bvp(ode, bcs, T, Y)
y_plot = solution.sol(T)
plt.title("Measles")
plt.plot(T, y_plot[0], label = "Susceptible")
plt.plot(T, y_plot[1], label = "Exposed")
plt.plot(T, y_plot[2], label = "Infectious")
plt.legend(loc="best")
plt.xlabel("Days (T)")
plt.ylabel("Percentage of Population")
plt.show()
```
