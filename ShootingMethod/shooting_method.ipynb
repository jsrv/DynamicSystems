```python
import numpy as np
from scipy.integrate import odeint
from matplotlib import pyplot as plt
```

# The Shooting Method for Boundary Value Problems

### Example of Secant Method

```python
# y'' +4y = -9sin(x), y(0) = 1., y(3*pi/4.) = -(1.+3*sqrt(2))/2., y'(0) = -2
# Exact Solution: y(x) = cos(2x) + (1/2)sin(2x) - 3sin(x)
def find_t(f,a,b,alpha,beta,t0,t1,maxI):
    sol1 = 0
    i = 0
    while abs(sol1-beta) > 10**-8 and i < maxI:
        sol0 = odeint(f, np.array([alpha,t0]), [a,b], atol=1e-10)[1,0]
        sol1 = odeint(f, np.array([alpha,t1]), [a,b], atol=1e-10)[1,0]
        t2 = t1 - (sol1 - beta)*(t1-t0)/(sol1-sol0)
        t0 = t1
        t1 = t2
        i = i+1
    if i == maxI:
        print("t not found")
    return t2

def solveSecant(f,X,a,b,alpha,beta,t0,t1,maxI):
    t = find_t(f,a,b,alpha,beta,t0,t1,maxI)
    sol = odeint(f,np.array([alpha,t]), X,atol=1e-10)[:,0]
    return sol

def ode(y,x):
    return np.array([y[1], -4*y[0]-9*np.sin(x)])

X = np.linspace(0,3*np.pi/4,100)
Y = solveSecant(ode,X,0,3*np.pi/4,1,-(1.+3*np.sqrt(2)) /2,
(1+(1.+3*np.sqrt(2)) /2)/(-3*np.pi/4),-1,40)
plt.plot(X,Y,'-k',linewidth=2)
plt.title("Example previously coded")
plt.show()
```

### Secant Method App

```python
alpha_ = 1
beta_  = 1
t_not  = (beta_-alpha_)/(np.pi)
tosel = np.copy(t_not)
def find_t(f,a,b,alpha,beta,t0,t1,maxI):
    sol1 = 0
    i = 0
    while abs(sol1-beta) > 10**-8 and i < maxI:
        sol0 = odeint(f, np.array([alpha,t0]), [a,b], atol=1e-10)[1,0]
        sol1 = odeint(f, np.array([alpha,t1]), [a,b], atol=1e-10)[1,0]
        t2 = t1 - (sol1 - beta)*(t1-t0)/(sol1-sol0)
        t0 = t1
        t1 = t2
        i = i+1
    if i == maxI:
        print("t not found")
    return t2

def solveSecant(f,X,a,b,alpha,beta,t0,t1,maxI):
    t = find_t(f,a,b,alpha,beta,t0,t1,maxI)
    sol = odeint(f,np.array([alpha,t]), X,atol=1e-10)[:,0]
    return sol

def ode(y,x):
    return np.array([y[1], -4*y[0]-9*np.sin(x)])

X = np.linspace(0,np.pi,100)
Y = solveSecant(ode,X,0,np.pi,alpha_,beta_,t_not,3,40)
plt.plot(X,Y,'-b',linewidth=2)
X = np.linspace(0,np.pi,100)
Y = solveSecant(ode,X,0,np.pi,alpha_,beta_,t_not,-1,40)
plt.plot(X,Y,'-g',linewidth=2)
plt.title("Using "+str(tosel)+" as our initial value")
plt.show()
```

### Newton Method Implementation

```python
alpha_ = 6
beta_  = np.exp(2)+6/np.exp(1)
t_not  = (beta_-alpha_)/(np.pi)

def find_tt(f,a,b,alpha,beta,t0,t1,maxI):
    sol = np.array([np.inf,0,0,0])
    i = 0
    while abs(sol[0]-beta) > 10**-8 and i < maxI:
        sol = odeint(f, np.array([alpha,t0,0,1]), [a,b], atol=1e-10)[1,:]
        t1 = t0 - (sol[0] - beta)/(sol[2])
        t0 = t1
        i = i+1
    if i == maxI:
        print("t not found")
    return t1

def solveNewton(f,X,a,b,alpha,beta,t0,t1,maxI):
    t = find_tt(f,a,b,alpha,beta,t0,t1,maxI)
    sol = odeint(f,np.array([alpha,t,0,1]), X,atol=1e-10)[:,0]
    return sol

def ode(y,x):
    return(np.array([y[1], 3+(2*y[0])/(x**2), y[3], (2/(x**2))*y[2]]))

X_ = np.linspace(1, np.e, 100)
Y = solveNewton(ode,X_,1, np.e, alpha_,beta_,t_not,-1,40)
plt.plot(X,Y,'-g',linewidth=2)
plt.title("Newton's method to solve the BVP")
plt.show()
print("Our initial guess was "+str(t_not))
```

### Cannon Shooting

```python
g_ = 9.8067
alpha_s = 0
omega_s = 0
beta_s = 0
v_ = 45

def find_t3(f, a, b, alpha,v_, beta, t0, t1, maxI):
    sol1 = np.inf
    i = 0
    while abs(sol1-beta) > 10**-8 and i < maxI:
        sol0 = odeint(f, np.array([alpha_s, v_, t0]), [a,b], atol=1e-10)[1,0]
        sol1 = odeint(f, np.array([alpha_s, v_, t1]), [a,b], atol=1e-10)[1,0]
        t2 = t1 - (sol1 - beta)*(t1-t0)/(sol1-sol0)
        t0 = t1
        t1 = t2
        i = i+1
    if i == maxI:
        print("t not found")
    return t2

def solveSecant(f,X,a,b,alpha,v_,beta,t0,t1,maxI):
    t = find_t3(f,a,b,alpha,v_,beta,t0,t1,maxI)
    sol = odeint(f,np.array([alpha,v_,t]), X, atol=1e-10)[:,0]
    return sol

miu1 = 0.0003
miu2 = 0

def ode2(y,x):
    return(np.array([np.tan(y[2]), -1*(g_*np.sin(y[2])+miu2*y[1]**2)/(y[1]*np.cos(y[2])), -g_/y[1]**2]))

def ode1(y,x):
    return(np.array([np.tan(y[2]), -1*(g_*np.sin(y[2])+miu1*y[1]**2)/(y[1]*np.cos(y[2])), -g_/y[1]**2]))

X_ = np.linspace(0, 195, 100)

Y1 = solveSecant(ode1, X_, 0, 195, alpha_s, v_, beta_s, 100, -5, 40)
plt.plot(X_,Y1,'--b',linewidth=2, label='miu = 0.0003')
Y2 = solveSecant(ode1, X_, 0, 195, alpha_s, v_, beta_s, np.e, 20*np.e, 40)
plt.plot(X_,Y2,'-b',linewidth=2, label='       = 0.0003')

Y3 = solveSecant(ode2, X_, 0, 195, alpha_s, v_, beta_s, 10, -.05, 40)
plt.plot(X_,Y3,'--g',linewidth=2, label='miu = 0')
Y4 = solveSecant(ode2, X_, 0, 195, alpha_s, v_, beta_s, 15, -10, 40)
plt.plot(X_,Y4,'-g',linewidth=2, label='       = 0')

plt.xlim(0,195)
plt.ylim(-10,75)
plt.title("Shooting Equidistant Cannon | Different Resistance")
plt.legend(loc='lower center')
plt.show()
print()
```
